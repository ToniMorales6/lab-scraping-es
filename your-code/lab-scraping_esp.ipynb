#### Desafío 10 - La información de los 20 últimos terremotos (fecha, hora, latitud, longitud y nombre de la región) por el EMSC como un dataframe de pandas

````python
# This is the url you will scrape in this exercise
#url7 = 'https://www.emsc-csem.org/Earthquake/'
url10 = "https://ds.iris.edu/seismon/eventlist/index.phtml"
tenlang = requests.get(url10).text
soup10 = BeautifulSoup(tenlang,"html.parser")
````

````python
data = [] #this is going to be a list of lists, each element will be the info of a earthquake
table = soup10.find_all('tr')
table
````

````python
# Encontramos todas las filas de la tabla
rows = soup10.find_all('tr')

# Iteramos sobre cada fila y extraemos los datos
for td in table:
    cells = td.find_all('td')
    if len(cells) == 0:  # Saltamos las filas sin datos si hubiese
        continue
    
    date_time_text = cells[0].text.strip()
    latitude_text = cells[1].text.strip()
    longitude_text = cells[2].text.strip()
    magnitude_text = cells[3].text.strip()
    depth_text = cells[4].text.strip()
    region_text = cells[5].text.strip()
    data.append([date_time_text, latitude_text, longitude_text, magnitude_text, depth_text, region_text])

terremotos = pd.DataFrame(data,columns=['date', 'latitude', 'longitude', "magnitude", "depth", "region"])
terremotos.head()
````

#### Desafío 11 - Datos del Top 250 de IMDB (nombre de la película, lanzamiento inicial, nombre del director y estrellas) como un dataframe de pandas

````python
# This is the url you will scrape in this exercise 
url11 = 'https://www.filmaffinity.com/es/userlist.php?user_id=882839&list_id=1006'
pelis = requests.get(url11)
soup11 = BeautifulSoup(pelis.content,"html.parser")
soup11
````

````python
# Tu código aquí
imbd = []
tags = list(soup11.select('div.mc-title > a'))
titles = [link.text for link in tags]

year = list(soup11.select('div.d-flex > span'))
yearnb = [link.text for link in year]
rating = list(soup11.select('div.mr-rating > div.avgrat-box'))
ratingnb = [link.text for link in rating]
director = list(soup11.select('div.mc-director > div > span > a'))
directornm = [link.text for link in director]

for tag, year, rating, director in zip(tags, yearnb, ratingnb, directornm):
    # Crear un diccionario para cada fila
    row_dict = {
        'titles': tag.text.strip(),
        'year': year.strip(),
        'rating': rating.strip(),
        'director': director.strip()
    }
    # Agregar el diccionario a la lista
    imbd.append(row_dict)

top250 = pd.DataFrame(imbd,columns=['titles', 'year', 'rating', "director"])
top250.head()
````
